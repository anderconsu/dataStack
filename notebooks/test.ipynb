```python
#SEPARAMOS LA DATA PARA SU ENTRENAMIENTO
X = casas_dummies.drop('price', axis=1)
y = casas_dummies['price']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
#IMPORTAMOS EL MODELO
from sklearn.ensemble import RandomForestRegressor
```

```python
#UTILIZAREMOS UN MODELO DE RANDOM FOREST CLASSIFIER
rnd_re = RandomForestRegressor(n_estimators=100, max_depth= 15)
```

```python
#UTILIZAREMOS UN MODELO DE RANDOM FOREST CLASSIFIER
rnd_re = RandomForestRegressor(n_estimators=100, max_depth= 15)
#ENTRENAMOS EL MODELO
rnd_re.fit(X_train, y_train)
```

```python
#PREDECIMOS
y_pred_RFC = rnd_re.predict(X_test)
```

```python
y_pred_RFC
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#PREDECIMOS
y_pred_RFC = rnd_re.predict(X_test)
# Calcular el Error Cuadrático Medio (MSE)
mse = mean_squared_error(y_test, y_pred_RFC)

# Calcular el Error Absoluto Medio (MAE)
mae = mean_absolute_error(y_test, y_pred_RFC)

# Calcular el Coeficiente de Determinación (R^2)
r2 = r2_score(y_test, y_pred_RFC)

# Imprimir las métricas
print(f"Error Cuadrático Medio (MSE): {mse}")
print(f"Error Absoluto Medio (MAE): {mae}")
print(f"Coeficiente de Determinación (R^2): {r2}")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {accuracy}")

# Mostrar la matriz de confusión y el informe de clasificación
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Matriz de Confusión:")
print(conf_matrix)

print("\nInforme de Clasificación:")
print(class_report)
```
